### for loop 사용
- 데이터 추출
- 텍스트만 동시 추출 및 공백 동시 제거

```python
cnt = 1
# cnt : 1로 초기화, 1부터 카운터가 되도록 세팅

for i in range(len(reviews)):
    print(f'[{cnt}]') # 리뷰보기 편하게 숫자로 출력구분
    print(reviews[i].text.strip()) # 공백제거하여 출력
    print() # 결과 보기 좋게 개행 추가
    cnt += 1
```

### 1~5페이지의 review data 수집

```python
# Selenium 자동브라우저가 아닌, beautifulsoup으로 진행
# key point는 url의 로직!

'https://review7.cre.ma/oneroommaking.com/products/reviews?app=0&iframe=1&iframe_id=crema-product-reviews-2&parent_url=https%3A%2F%2Foneroommaking.com%2Fproduct%2Fdetail.html%3Fproduct_no%3D1911%26cate_no%3D61%26display_group%3D1&product_code=1911&secure_device_token'
'https://review7.cre.ma/oneroommaking.com/products/reviews?app=0&iframe=1&iframe_id=crema-product-reviews-2&page=2&parent_url=https%3A%2F%2Foneroommaking.com%2Fproduct%2Fdetail.html%3Fproduct_no%3D1911%26cate_no%3D61%26display_group%3D1&product_code=1911&secure_device_token=V2c26c49798b678b5367dde9393aefc632a7ec4b849d05080fd331cef21fd252c8bfaf2e281921deb066e40748c3824d48&widget_env=100&widget_style='
'https://review7.cre.ma/oneroommaking.com/products/reviews?app=0&iframe=1&iframe_id=crema-product-reviews-2&page=3&parent_url=https%3A%2F%2Foneroommaking.com%2Fproduct%2Fdetail.html%3Fproduct_no%3D1911%26cate_no%3D61%26display_group%3D1&product_code=1911&secure_device_token=V2c26c49798b678b5367dde9393aefc632a7ec4b849d05080fd331cef21fd252c8bfaf2e281921deb066e40748c3824d48&widget_env=100&widget_style='
# 첫번째 페이지는 제품 리뷰를 보여주는 위젯을 처음 로드
# 두번째, 세번째 페이지 부터 로딩된 위젯 내에서 페이징(페이지 왔다갔다), 다음페이지의 리뷰 로딩
```

```python
# for loop 사용해서 1~5페이지 리뷰 가져오기
for pageNo in range(1,6): # pageNo : 1,2,3,4,5
    url = f'https://review7.cre.ma/oneroommaking.com/products/reviews?app=0&iframe=1&iframe_id=crema-product-reviews-2&page={pageNo}&parent_url=https%3A%2F%2Foneroommaking.com%2Fproduct%2Fdetail.html%3Fproduct_no%3D1911%26cate_no%3D61%26display_group%3D1&product_code=1911&secure_device_token=V2c26c49798b678b5367dde9393aefc632a7ec4b849d05080fd331cef21fd252c8bfaf2e281921deb066e40748c3824d48&widget_env=100&widget_style='
    # url : 1페이지는 page부분이 생략되어 있음 (2페이지 이후부터 아무거나 복사해서 페이지부분 포맷팅)
    
    res = req.get(url) # get요청
    print(f"[현재 페이지 번호 : {pageNo}]") # 현재 페이지 번호
    soup = bs(res.text, 'lxml') # beautifulsoup 객체화
    reviews = soup.select('div.review_list_v2__message.js-collapsed-review-content.js-translate-text')
    ############################여기까지가 데이터 추출 과정###########################################################
    
    # for loop을 하나 더 사용하여 공백 제거 (strip()함수 사용)
    for i in range(len(reviews)):
        print(reviews[i].text.strip()) # 공백제거한 리뷰
        print() # 개행
    print('-'*80)
    # capture 또는 '로봇이 아닙니다' 클릭항목 = 사용자 인증 방법
    # 너무 많은 요청이 들어오면, 사용자 인증하는 방법으로 보호
    # time.sleep()으로 일정한 시간 간격을 주어서 웹페이지에 요청
    time.sleep(1) # 웹페이지가 바뀔 때의 시간을 확보해 주기 위해!!(적절한 딜레이 시간 설정)

print('수집 완료!')
```

```python
# params 매개변수를 사용하여 간단하게 '쿼리 스트링'을 추가하는 방법

# 'https://review4.cre.ma/bodyluv.kr/products/reviews?'
# url의 끝에 물음표(?)를 포함하여 요청 매개변수(request parameter)를 전달하기 위한 방법

for pageNo in range(1,6):
    print(f"[{pageNo}]")
    url2 = 'https://review7.cre.ma/oneroommaking.com/products/reviews?app=0&iframe=1&iframe_id=crema-product-reviews-2&parent_url=https%3A%2F%2Foneroommaking.com%2Fproduct%2Fdetail.html%3Fproduct_no%3D1911%26cate_no%3D61%26display_group%3D1&product_code=1911&secure_device_token=V2fc98f6259c654f126d62766e4ecf3f776c5f2b2aabadcc811e8b6e4ab1e67c01b3553921dea8a20d687711b910f76fc6&widget_env=100&widget_style=#'
    res2 = req.get(url2, params={'page':pageNo})
        # params : dictionary로 정의
        # key : url의 'page='의 page로 정의
        # value : page=3일 때의 3
        
        # FAQ : 첫번째 페이지에 page=1이 생략이 되는건가요?
            # 페이지 번호는 생략했지만 page=1이 디폴트로 설정되어 있는 것
    soup2 = bs(res2.text, 'lxml') # beautifulsoup 객체화
    reviews2 = soup2.select('div.review_list_v2__message.js-collapsed-review-content.js-translate-text')
    for i in range(len(reviews2)):
        print(reviews2[i].text.strip())
        print()
    print('-'*80)
    time.sleep(2)
print('수집완료!')
```

   - sql 쿼리문과 비슷.
   - sql문과 유사하게 '쿼리 스트링'을 사용해서 원하는 데이터를 추출/필터링 가능
   - 무조건 쓸어담는 게 아니라, 경계를 정해주고, 필요한 데이터를 세부적으로 뽑아올 수 있음.
### 파일 생성

```python
# 파일을 생성하고, 열고, 작성할거라는 선언
f = open('review.txt', 'w', encoding='utf-8')


for pageNo in range(1,6): # pageNo : 1,2,3,4,5
    url = f'https://review7.cre.ma/oneroommaking.com/products/reviews?app=0&iframe=1&iframe_id=crema-product-reviews-2&page={pageNo}&parent_url=https%3A%2F%2Foneroommaking.com%2Fproduct%2Fdetail.html%3Fproduct_no%3D1911%26cate_no%3D61%26display_group%3D1&product_code=1911&secure_device_token=V2c26c49798b678b5367dde9393aefc632a7ec4b849d05080fd331cef21fd252c8bfaf2e281921deb066e40748c3824d48&widget_env=100&widget_style='
    # url : 1페이지는 page부분이 생략되어 있음 (2페이지 이후부터 아무거나 복사해서 페이지부분 포맷팅)
    
    res = req.get(url) # get요청
    print(f"[현재 페이지 번호 : {pageNo}]") # 현재 페이지 번호
    soup = bs(res.text, 'lxml') # beautifulsoup 객체화
    reviews = soup.select('div.review_list_v2__message.js-collapsed-review-content.js-translate-text')
    ############################여기까지가 데이터 추출 과정###########################################################
    
    # for loop을 하나 더 사용하여 공백 제거 (strip()함수 사용)
    for i in range(len(reviews)):
        print(reviews[i].text.strip()) # 공백제거한 리뷰
        print() # 개행
        f.write(reviews[i].text.strip()+'\n')
    print('-'*80)
    
    
    # capture 또는 '로봇이 아닙니다' 클릭항목 = 사용자 인증 방법
    # 너무 많은 요청이 들어오면, 사용자 인증하는 방법으로 보호
    # time.sleep()으로 일정한 시간 간격을 주어서 웹페이지에 요청
    time.sleep(1) # 웹페이지가 바뀔 때의 시간을 확보해 주기 위해!!(적절한 딜레이 시간 설정)

print('수집 완료!')

# 파일 닫기
f.close()
```
