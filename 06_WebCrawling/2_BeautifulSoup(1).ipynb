# 파싱 (Parsing)이란?
- 웹페이지에서 원하는 데이터를 추출, 가공하기 쉬운 상태로 변환하는 것
- 예를 들면, 생선을 낚아서 최종적으로 통조림의 형태로 만들어야하는 데, 이 때의 손질하려는 절차

생선과는 달리, 웹페이지에 존재하는 데이터는 list, dictionary와 같은 자료구조와 다르다. <br>
그래서 parser함수나 프로그램으로 다루기 쉬운 형태로 바꾸어주는데 
이 과정을 '파싱'이라고 한다.

```python
# 응답받은 데이터(res_google)를 beautifulsoup을 사용해서 변환
bs_google = bs(res_google.text, 'lxml')
bs_google
```

```python
# 자료 유형 알아보기
print(type(res_google))
print(type(bs_google))

# beautifulsoup으로 객체화 진행 된 것 확인.
```
