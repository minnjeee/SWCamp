# Melon Chart top 100 수집[타이틀|아티스트]
- 타이틀과 아티스트명을 beautifulsoup을 사용해서 수집
- for loop 적용
- 마지막으로, 데이터를 pandas library의 dataframe에 집어넣기
- 데이터 저장 파일을 보내기

```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://www.melon.com/chart/'
res_chart = req.get(url)
res_chart # 클라이언트의 요청이 문제
```

```python
head = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}
res_chart = req.get(url, headers=head)
res_chart
```

```python
soup_chart = bs(res_chart.text, 'lxml')
soup_chart
```

```python
songs = soup_chart.select('div.rank01.ellipsis > span > a')
artists = soup_chart.select('div.rank02.ellipsis > span')
# artists는 a태그 설정 안한 이유 : 듀엣곡 때문(한 곡에 최소 2명의 아티스트), 즉 div태그안에 a태그가 두개였음
for i in range(len(songs)):
    print(songs[i].text, artists[i].text, sep=' - ')
    
# 크롤링은 결국 족보싸움!! 
# 내가 원하는 태그 내 데이터에 Id,Class가 불확실하다면, 그 태그의 상위태그까지 올라가야 할 수도 있음
```

### 수집한 타이틀,아티스트 데이터를 for loop을 통해 리스트에 집어넣자

```python
# 자료형을 확인해보자
type(songs) 
# bs4.element.ResultSet : 리스트와 비슷하지만 엄연히 다른 구조임!
# 우리는 python 기반 pandas library를 사용해야 하므로 빈 리스트 2개를 선언해서 for loop으로 순차적으로 담자
```

```python
# 빈 리스트 선언
song_list = []
artist_list = []

for i in range(len(songs)):
    song_list.append(songs[i].text)
    artist_list.append(artists[i].text)
    print(song_list[i], artist_list[i])
```

### DataFrame으로
1. Dictionary 형태로 만들어 줘야합니다.
    - Dictionary로 우선 선언 이유
      - 각각의 행과 열에 대한 데이터를 쉽게 입력 가능
      - frame 내 데이터 타입을 일치시키기 쉽게 구성 가능
      - dictionary로 먼저 저장을 해주면, 데이터 구조가 보존됨(데이터프레임으로 변경 시 일관성 유지됨)


```python
melon_dic = {'타이틀' : song_list, '아티스트':artist_list}
melon_dic
```

```python
# DataFrame
melon_df = pd.DataFrame(melon_dic)
melon_df
```

```python
# CSV파일로 내보내기
# (항상 파일로 내보내는 습관 필요)
melon_df.to_csv('C:/Users/user02/WebCrawling/melon_chart.csv', encoding='euc-kr')
```

# 실습 : SUV SPORTS 옵션 가져오기
### Library Import

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://www.cadillac.co.kr/crossovers-suvs/escalade'
res = req.get(url)
res # 성공적으로 가져옴
```

### BeautifulSoup 객체화

```python
# slect 등 beautifulsoup에서 제공하는 메소드와 함수를 사용하기 위해 필수!
bs_cadi = bs(res.text, 'lxml')
bs_cadi
```

```python
# CSS selector를 활용해서 차량명 뽑아보기
suvNames = bs_cadi.select('h4.q-headline.none-margin')
suvNames[1].text
```

#### \n, \t => 이스케이프 문자(Escape Character)
    - \n : 새로운 줄(enter)
    - \t : 간격(tab)
  - '\n\t\t\tSPORT PLATINUM\n\t\t' : 앞에는 3개의 탭, 뒤에는 두개의 탭과 새로운 줄이 있음

```python
# replace()함수를 사용, strip()함수 사용
for name in suvNames:
    suvSports = name.text.replace('\n','').replace('\t','').strip()
    # replace만 하면 끝인가? no no
    # replace는 공백을 제거한 것처럼 보이지만, 개형문자가 빈 문자로 대체될 뿐,
    # 실제로 공백이 완전히 제거되지 않았음! => strip()함수 사용
    print(suvSports)
```

### 옵션 추출

```python
features = bs_cadi.select('adv-grid.none-margin.grid-sm-jc-center > adv-col ~ adv-col  div.q-text.q-body1>ul>li')
feature = []
n=1
for i in features:
    print(f'특징{n} \t: {i.text.replace("1)","").strip()}')
    feature.append(i.text.replace("1)","").strip())
    n += 1
```

### DataFrame 
- 차랑명은 'Sports Platinum'
- 옵션들

```python
df_cadi = pd.DataFrame({'SUV_name':suvSports,'SUV_features':feature})
df_cadi
```

- 데이터 추출
