# 실습 : SUV SPORTS 옵션 가져오기
### Library Import

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://www.cadillac.co.kr/crossovers-suvs/escalade'
res = req.get(url)
res # 성공적으로 가져옴
```

### BeautifulSoup 객체화

```python
# slect 등 beautifulsoup에서 제공하는 메소드와 함수를 사용하기 위해 필수!
bs_cadi = bs(res.text, 'lxml')
bs_cadi
```

```python
# CSS selector를 활용해서 차량명 뽑아보기
suvNames = bs_cadi.select('h4.q-headline.none-margin')
suvNames[1].text
```

#### \n, \t => 이스케이프 문자(Escape Character)
    - \n : 새로운 줄(enter)
    - \t : 간격(tab)
  - '\n\t\t\tSPORT PLATINUM\n\t\t' : 앞에는 3개의 탭, 뒤에는 두개의 탭과 새로운 줄이 있음

```python
# replace()함수를 사용, strip()함수 사용
for name in suvNames:
    suvSports = name.text.replace('\n','').replace('\t','').strip()
    # replace만 하면 끝인가? no no
    # replace는 공백을 제거한 것처럼 보이지만, 개형문자가 빈 문자로 대체될 뿐,
    # 실제로 공백이 완전히 제거되지 않았음! => strip()함수 사용
    print(suvSports)
```

### 옵션 추출

```python
features = bs_cadi.select('adv-grid.none-margin.grid-sm-jc-center > adv-col ~ adv-col  div.q-text.q-body1>ul>li')
feature = []
n=1
for i in features:
    print(f'특징{n} \t: {i.text.replace("1)","").strip()}')
    feature.append(i.text.replace("1)","").strip())
    n += 1
```

### DataFrame 
- 차랑명은 'Sports Platinum'
- 옵션들

```python
df_cadi = pd.DataFrame({'SUV_name':suvSports,'SUV_features':feature})
df_cadi
```
